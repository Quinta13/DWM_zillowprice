## Rimozione colonne con molti Nan ##

```python
# Given the dataframe and the name of the column returns the column
def get_col(df, colName):
    return df.loc[:, colName]

# Given a column returns Nan-count and Nan-percentage
def get_col_nan_info(col):
    count = col.isna().sum()
    tot = len(col)
    perc = count/tot
    return count, perc

# Given the df and a treshold returns a list of column names with Nan-percentage greater or equal the treshold
def get_cols_over_nan_percentage(df, treshold):
    names = df.columns
    overPercentage = []
    for name in names:
        col = get_col(df, name)
        _ , perc = get_col_nan_info(col)
        # print(f'{name}: {perc})')
        if perc > treshold:
            overPercentage.append(name)
    return overPercentage

over_nan = get_cols_over_nan_percentage(df_2016, 0.7)

for o in over_nan:
    print(f'{o} : {get_col_nan_info(get_col(df_2016, o))}')
print(f'Legth: {len(over_nan)}')
```

```python
print(df_2016.shape)
df_2016 = df_2016.drop(over_nan, axis=1)
print(df_2016.shape)
```

## Rimozione righe con molti Nan ##

```python
# Given the dataframe and the index of the row returns the row
def get_row(df, index):
    return df.loc[index, :]

# Given a row returns Nan-count and Nan-percentage
def get_row_nan_info(row):
    count = row.isna().sum()
    tot = len(row)
    perc = count/tot
    return count, perc

# Given the df and a treshold returns a list of row ids with Nan-percentage greater or equal the treshold
def get_rows_over_nan_percentage(df, idName, treshold):
    overPercentage = []
    overPercentage_indexes = []
    for i in df.index:
        row = get_row(df, i)
        _ , perc = get_row_nan_info(row)
        if perc > treshold:
            overPercentage.append(row[idName])
            overPercentage_indexes.append(i)
    return overPercentage, overPercentage_indexes

over_nan, indexes = get_rows_over_nan_percentage(df_2016, "parcelid", 0.5)

for o, i in zip(over_nan, indexes):   
    print(f'{o} : {get_row_nan_info(get_row(df_2016, i))}')
print(f'Legth: {len(over_nan)}')
```

## Controllo se i valori sono discreti ##

```python
def get_discrete(df, treshold):
    discretes = []
    for col_name in df.columns:
        values_count = len(get_col(df,col_name).unique())
        if values_count < treshold:
            discretes.append(col_name)
    return discretes

discretes = get_discrete(df_2016, 30)
print(f'{discretes} ({len(discretes)})')
```

```python
for discrete in discretes:
    values = get_col(df_2016, discrete).unique()
    print(f'{discrete} : {values} ({len(values)})')
```

## RIEMPIRE I nan ##

PRIMA: aggiungere colonne con 1 se c'è valore, 0 se c'è Nan

```python
numerics = list(set(df_2016.columns) - set(discretes + not_numeric))
numerics = numerics + ordinal
print(numerics)
```

```python
def fill_nan_with_mean(df, col_names):
    # add a column that keeps orginal nan values
    df.loc[:,'NanCount'] = df[col_names].isna().astype(int).sum(axis=1)
    for col_name in col_names:
        col = get_col(df, col_name)
        col.fillna(value=col.mean(), inplace=True)
    
fill_nan_with_mean(df_2016, numerics)
```

## One-Hot encoding ##

```python
from sklearn.preprocessing import OneHotEncoder
def one_hot_encoding(df, col_names):
    oh = OneHotEncoder(sparse=False)
    oh.fit(df[col_names])
    encoded = oh.transform(df[col_names])
    for i,col in enumerate(oh.get_feature_names()):
        df[col] = encoded[:,i]
        
one_hot_encoding(df_2016, categorical)
```
